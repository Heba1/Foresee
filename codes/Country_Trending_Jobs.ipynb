```python
import pandas as pd
from pandas import DataFrame as df
from collections import Counter
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"jobDataset_wadhefa_cleaned.csv")
```

```python
df.job_title.value_counts().nlargest(10) #Top Ten jobs of all years
```

```python
mask = df['job_date'].isin([2018])  # Mask to select only job date that equal 2018

mask
```

```python
df[mask]  # get all jobs titles that related to 2018
```

```python
df[mask].job_title.value_counts().nlargest(10).sort_index().plot('bar')  # Show Trending job of all years ago 

plt.xlabel('Job')
plt.ylabel('Growth')
plt.title('Trending Ten Jobs Of 2018\n')
plt.savefig("Trending_Ten_Jobs_Of_2018.png")
```

```python
df.job_title.value_counts().nlargest(10).sort_index().plot('bar')   # Show Trending job at 2018 

plt.xlabel('Job')
plt.ylabel('Growth')
plt.title('Trending Ten Jobs Of All Years\n')
plt.savefig("Trending_Ten_Jobs_Of_All_Years.png")
```
