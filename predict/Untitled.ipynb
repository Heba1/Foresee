```python
import pandas as pd 
from prediction import prediction_Class




```

```python

dataset=pd.read_csv("../data/jobDataset_All.csv")
dataset=dataset[dataset.job_specialty=="Accounting And Auditing"]
```

```python
dataset
```

```python
ma=dict()
for i in range(len(dataset)):
    x=dataset.iloc[i]
    if not x.job_date in ma:
        ma.update({x.job_date:1})
    else:
        ma[x.job_date]+=1

```

```python
date=list()
count=list()

for key, value in ma.items():
        date.append(key)
        count.append(value)
```

```python
n_data=pd.DataFrame({"date":date,"count":count})
```

```python
n_data
```

```python
pre_cols=["date"]
target="count"

p =prediction_Class(n_data,predic_columns=pre_cols,targe=target)
```
